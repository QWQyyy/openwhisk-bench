```python
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results_dir = "prime_number"
input = '1000'
requests = '2000'
files = os.listdir(results_dir)

cpuMap = {}
ramMap = {}

for file in files:
    tokens = file.split("_")
    cpuList = []
    ramList = []
    if tokens[1] == input and tokens[2] == requests:
        lines = [line.rstrip('\n') for line in open("{0}/{1}".format(results_dir, file))]
        for line in lines:
            params = line.split(" ")
            cpuList.append(int(params[0]))
            ramList.append(int(params[1]))
        cpuMap[tokens[0]] = cpuList
        ramMap[tokens[0]] = ramList
        
    
    
    
```

```python
plt.figure(figsize=[25, 12])
#plt.xticks(labels)
invokerDataToPrint = ['1', '2', '3', '4', '5', '6', '7']
for e in invokerDataToPrint:
    plt.plot(cpuMap[e], label='Invoker #{0}: CPU'.format(e))
plt.ylabel('CPU Load, %', fontsize = 24)
plt.xlabel('Time, seconds', fontsize = 24)
plt.rc('xtick', labelsize=24) 
plt.rc('ytick', labelsize=24) 
plt.legend(prop={'size': 24})
plt.show()
```
