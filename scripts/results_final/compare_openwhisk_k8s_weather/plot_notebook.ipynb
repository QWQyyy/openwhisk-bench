```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
labels = np.array([50, 100, 500, 1000, 2000])

openwhisk = np.array([5.89939713478, 5.47983002663, 33.6919720173, 69.1840660572, 123.932620049])
k8s = np.array([1.09451603889, 1.18463897705, 3.35382294655, 2.58712506294, 4.27715206146])
```

```python
plt.figure(figsize=[10, 5])
#plt.xticks(labels)
plt.plot(labels, openwhisk, label='Openwhisk')
plt.plot(labels, k8s, label='Spring Boot')
plt.xlabel('Requests')
plt.ylabel('Duration, seconds')
plt.legend()
plt.show()
```
